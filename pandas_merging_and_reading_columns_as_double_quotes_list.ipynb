**Adding break tag**

```python
data['Text'] = data['Text'].str.replace(r',', ',</br>')
```

**Removing specific speacila charaters from csv file**

```python
data['Text']= data['Text'].str.replace(r'[^A-Za-z0-9,</>]+', '')
data['Text'].to_csv('header.csv', index=False)
```

**Removing all special characters**

```python
import re
p = re.compile(r'[^\w\s]+')
data['Text'] = [p.sub('', x) for x in data['Text'].tolist()]
data['Text'].to_csv('header1.csv')
```

```python
import string

punct = '!"#$%&\'()*+-./:;<=>?@[\\]^_`{}~'   # `|` is not present here
transtab = str.maketrans(dict.fromkeys(punct, ''))

data['Text'] = '|'.join(data['Text'].tolist()).translate(transtab).split('|')
data['Text'].to_csv('header1.csv')
```

```python
import pandas as pd
import csv
with open('data.csv') as f:
    reader = csv.reader(f)
    with open('output.csv', 'w') as g:
        writer = csv.writer(g)
        for row in reader:
            new_row = [' '.join([row[1], row[2]])] + row[3:]
            writer.writerow(new_row)
```

**Reading csv as a List with double quotes and single quotes**

```python
import json
import pandas as pd

filedata = "data.csv"
dfjson = pd.read_csv(filedata, index_col=None, header=0)
input_list = dfjson['URL'].str.strip().values.tolist()

# Convert the list to a JSON-formatted string with double quotes
broken_data = json.dumps(input_list)
print(broken_data)
```

```python
import json

# Assuming dfjson['URL'].values.tolist() gives you ['link1', 'link2']
input_list = dfjson['URL'].str.strip().values.tolist()

# Convert the list to a JSON-formatted string
json_string = json.dumps(input_list)
print(json_string)
# json_string = json.dumps(input_list, ensure_ascii=False)
# Parse the JSON string back into a list
output_list = json.loads(json_string)

print(output_list)
```
